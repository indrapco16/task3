### Car Task

## Q.1 Load the dataset and print first 10 rows.		


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\HP\Desktop\New folder\BMW Car Task\BMW Car Task\bmw_car_data.csv")
```

```python
df.head(10)
```

## Q.2 Display shape, column names, and data types.		


```python
print(df.shape)
print(df.columns)
print(df.dtypes)
```

## Q.3 Check for missing values and count per column.		


```python
df.isnull().sum()
```

## Q.4 Filter cars manufactured after 2020.		


```python
df[df['Year'] > 2020]
```

## Q.5 List unique car models and fuel types.		


```python
print(df['Model_Name'].unique())
print(df['Fuel_Type'].unique())
```

## Q.6 Count how many cars per fuel type.		


```python
df['Fuel_Type'].value_counts()
```

## Q.7 Sort dataset by Price_Ex_Showroom descending.		


```python
df.sort_values('Price_Ex_Showroom', ascending=False)
```

## Q.8 Filter only Automatic transmission BMWs.		


```python
df[df['Transmission']=='Automatic']
```

## Q.9 Replace missing mileage values with mean.		


```python
df['Mileage_kmpl'].fillna(df['Mileage_kmpl'].mean(), inplace=True)
```

## Q.10 Drop rows where Engine_CC is null.		


```python
df.dropna(subset=['Engine_CC'], inplace=True)
```

## Q.11 Group cars by Model_Name and calculate average price.		


```python
df.groupby('Model_Name')['Price_Ex_Showroom'].mean()
```

## Q.12 Create a new column Car_Age = 2025 - Year.		


```python
df['Car_Age'] = 2025 - df['Year']
```

## Q.13 Filter cars priced above ₹60 lakh.		


```python
df[df['Price_Ex_Showroom'] > 60]
```

## Q.14 Find top 5 most common cities in the dataset.		


```python
df['Location'].value_counts().head(5)
```

## Q.15 Plot histogram for mileage.		


```python
df['Mileage_kmpl'].hist()
plt.show()
```

## Q.16 Create a bar chart of fuel type vs average power.		


```python
df.groupby('Fuel_Type')['Power_BHP'].mean().plot(kind='bar')
plt.show()
```

## Q.17 Apply .loc[] to update fuel type of electric variants.		


```python
df.loc[df['Variant']=='xDrive','Fuel_Type']='Electric'
```

## Q.18 Convert Price_Ex_Showroom from lakh to ₹ value.		


```python
df['Price_Rupees'] = df['Price_Ex_Showroom'] * 100000
```

## Q.19 Find how many cars are older than 10 years.		


```python
(df['Car_Age'] > 10).sum()
```

## Q.20 Count of cars by Owner_Type and Transmission.		


```python
df.groupby(['Owner_Type','Transmission']).size()
```

 ## Q.21 Find maximum BHP in diesel cars.		


```python
df[df['Fuel_Type']=='Diesel']['Power_BHP'].max()
```

## Q.22 Plot scatter plot of Engine_CC vs Price.		


```python
plt.scatter(df['Engine_CC'], df['Price_Ex_Showroom'])
plt.show()
```

## Q.23 Count number of variants per model.		


```python
df.groupby('Model_Name')['Variant'].count()
```

```python
## Q.24 Find car with max and min mileage.		
df.loc[df['Mileage_kmpl'].idxmax()]
df.loc[df['Mileage_kmpl'].idxmin()]
```

## Q.25 Calculate median price per year.		


```python
df.groupby('Year')['Price_Ex_Showroom'].median()		
```

## Q.26 Pivot table with Location vs Fuel_Type.		


```python
pd.pivot_table(df,index='Location',columns='Fuel_Type',values='Price_Ex_Showroom')
```

## Q.27 Create a column Is_Luxury (if Price > ₹70L = Yes).		


```python
df['Is_Luxury'] = df['Price_Ex_Showroom'].apply(lambda x:'Yes' if x>70 else 'No')
```

## Q.28 Remove duplicate rows if any.		


```python
df.drop_duplicates(inplace=True)
```

## Q.29 Apply filter: Diesel + Automatic + Above 200 BHP.		


```python
filtered = df[(df['Fuel_Type']=='Diesel') &
              (df['Transmission']=='Automatic') &
              (df['Power_BHP']>200)]
filtered
```

## Q.30 Export filtered data to new CSV.		


```python
filtered.to_csv("C:/Users/HP/Desktop/New folder/BMW Car Task/BMW Car Task/filtered_bmw.csv", index=False)
```
