### Q.3 Sales Trend Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel(r"C:\Users\HP\Desktop\sakshi 29\Retail Data Task Sheet\Retail Data\Indian_Retail_Sales_Dataset.xlsx")

# Convert Order Date to Date Format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract Year and Month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

# Monthly Sales Trend
monthly_sales = df.groupby('Month')['Sales'].sum()
print("Monthly Sales Trend:")
print(monthly_sales)

# Monthly Profit Trend
monthly_profit = df.groupby('Month')['Profit'].sum()
print("\nMonthly Profit Trend:")
print(monthly_profit)

# Yearly Sales Trend
yearly_sales = df.groupby('Year')['Sales'].sum()
print("\nYearly Sales Trend:")
print(yearly_sales)

# Sales Line Chart
monthly_sales.plot(kind='line', figsize=(10,5), marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

### Q.4 Category and Product Insights


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel(r"C:\Users\HP\Desktop\sakshi 29\Retail Data Task Sheet\Retail Data\Indian_Retail_Sales_Dataset.xlsx")

# Category-wise Sales
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print("Category-wise Sales:")
print(category_sales)

# Category-wise Profit
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print("\nCategory-wise Profit:")
print(category_profit)

# Sub-Category-wise Sales
subcategory_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print("\nTop Sub-Categories by Sales:")
print(subcategory_sales.head(10))

# Top 10 Products by Sales
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)
print("\nTop 10 Products by Sales:")
print(top_products.head(10))

# Category Sales Chart
category_sales.plot(kind='bar', figsize=(8,5))
plt.title("Category-wise Sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

### Q.5 Regional Performance


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel(r"C:\Users\HP\Desktop\sakshi 29\Retail Data Task Sheet\Retail Data\Indian_Retail_Sales_Dataset.xlsx")

# Region-wise Sales and Profit
region_performance = df.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

print("Region-wise Performance:")
print(region_performance)

# State-wise Sales
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

print("\nTop 10 States by Sales:")
print(state_sales.head(10))

# State-wise Profit
state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending=False)

print("\nTop 10 States by Profit:")
print(state_profit.head(10))

print("\nBottom 10 States by Profit:")
print(state_profit.tail(10))

# Region-wise Sales Chart
region_performance['Sales'].plot(kind='bar', figsize=(8,5))
plt.title("Region-wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()

# Region-wise Profit Chart
region_performance['Profit'].plot(kind='bar', figsize=(8,5))
plt.title("Region-wise Profit")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```

### Q.6 Customer Segmentation


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_excel(r"C:\Users\HP\Desktop\sakshi 29\Retail Data Task Sheet\Retail Data\Indian_Retail_Sales_Dataset.xlsx")

# Segment-wise Sales and Profit
segment_analysis = df.groupby('Segment')[['Sales', 'Profit']].sum()

print("Segment-wise Sales and Profit:")
print(segment_analysis)

# Top 10 Customers by Sales
top_customers_sales = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False)

print("\nTop 10 Customers by Sales:")
print(top_customers_sales.head(10))

# Top 10 Customers by Profit
top_customers_profit = df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False)

print("\nTop 10 Customers by Profit:")
print(top_customers_profit.head(10))

# Segment-wise Sales Chart
segment_analysis['Sales'].plot(kind='bar', figsize=(8,5))
plt.title("Segment-wise Sales")
plt.xlabel("Segment")
plt.ylabel("Sales")
plt.show()

# Segment-wise Profit Chart
segment_analysis['Profit'].plot(kind='bar', figsize=(8,5))
plt.title("Segment-wise Profit")
plt.xlabel("Segment")
plt.ylabel("Profit")
plt.show()
```

### Q.8 Shipping Performance


```python
import pandas as pd

df = pd.read_excel(r"C:\Users\HP\Desktop\sakshi 29\Retail Data Task Sheet\Retail Data\Indian_Retail_Sales_Dataset.xlsx")

# Date Conversion
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Delivery Time
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Average Delivery Time
print("Average Delivery Days:", df['Delivery Days'].mean())

# Maximum Delay
print("Maximum Delivery Days:", df['Delivery Days'].max())

# Minimum Delivery Days
print("Minimum Delivery Days:", df['Delivery Days'].min())
```
